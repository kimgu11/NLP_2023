```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X_train = np.random.rand(100, 1).astype(np.float32)
y_train = 2 * X_train + 1 + 0.1 * np.random.rand(100, 1).astype(np.float32)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_dim=1)
])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(X_train, y_train, epochs=100, verbose=1)

X_test = np.array([[0.2], [0.5], [0.8]], dtype=np.float32)
predictions = model.predict(X_test)

for i in range(len(X_test)):
    print(f"Input: {X_test[i][0]}, Predicted Output: {predictions[i][0]}")

plt.scatter(X_train, y_train, label='Training Data')
plt.plot(X_test, predictions, label='Predictions', color='red', marker='o')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
